```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style("white")
```

# Data Distributions

## 1 Normal Distribution (Gaussian)



* Bell-shaped density curve
* Parameters: mean and standard deviation (std)
* Symmetrical
* Mean-centered
* Standard deviation determines the spread

```python
# generate random numbers from a Normal distribution
m = 0
std = 1 
normal_sample = np.random.normal(loc=m, scale=std, size=1000)
```

```python
# plot data with seaborn
sns.histplot(normal_sample, 
             color='#f67280',
             kde=True,
            stat='density')
plt.title('Normal Distribution', fontsize=15)
plt.show()
```

## 2 Uniform Distribution

* Every data point has equal probability
* Parameters: start and width/end

```python
# generate a random sample from a Uniform distribution
start = 10
end = 20
uniform_sample = np.random.uniform(low=start, high=end, size=1000)
```

```python
# plot data with seaborn
sns.histplot(uniform_sample, 
             color='#f67280',
             kde=True,
             stat='density')
plt.title('Uniform Distribution', fontsize=15)
plt.show()
```

## 3 Binomial Distribution

* Models the probability of getting exactly k successes in n independent **Bernoulli trials**
* **Bernoulli distribution** is a discrete distribution consisting of only one trial with 2 outcomes (success/failure)
* Parameters: n - total number of trials; p - probability of success

```python
# generate a random sample from a Binomial distribution
binomial_sample = np.random.binomial(n=10, p=0.5, size=1000)
```

```python
# plot data with seaborn
sns.histplot(binomial_sample, 
             color='#f67280',
             kde=True,
             stat='density')
plt.title('Binomial Distribution', fontsize=15)
plt.show()
```

 ## 4 Poisson Distribution

* Models the rate at which an event occurs
* Parameters: Lambda (expected value or mean) - event rate
* If the number of events in a time interval follows a Poisson distribution, then the amount of time between events follows the exponential distribution.


```python
# generate a random sample from a Poisson distribution
poisson_sample = np.random.poisson(lam=10, size=1000)
```

```python
# plot data with seaborn
sns.histplot(poisson_sample, 
             color='#f67280',
             kde=True,
             stat='density')
plt.title('Poisson Distribution', fontsize=15)
plt.show()
```

## 5 Exponential Distribution

* Models 
* Parameters: Lambda - event rate parameter (same as in Poisson D)
* Decay parameter = 1/Lambda.
* Mean - 1/Lambda
* Continuous
* Helps to understand the amount of time until an event occures
* “Memoryless” property - time has no effect on future outcomes.
* If the number of events in a time interval follows a Poisson distribution, then the amount of time between events follows the exponential distribution.

```python
lam = 1
exponential_sample = np.random.exponential(scale=lam, size=1000)
```

```python
# plot data with seaborn
sns.histplot(exponential_sample, 
             color='#f67280',
             kde=True,
             stat='density')
plt.title('Exponential Distribution', fontsize=15)
plt.show()
```

## 6 Beta Distribution

* Often seen in Bayesian inference and order statistics
* It's a probability distribution on probabilities (bounded between 0 and 1)
* Models the probability of success of an event (for which the number of all trials and successful outcomes has been observed)
* Continuous
* Shape parameters: a - number of times the event happened; b - number of times the event didn't happen
* P is a random variable and denotes the probability of success. This corresponds to our different hypotheses for the possible probabilities.

```python
a = 5
b = 1195
beta_sample = np.random.beta(a=a, b=b, size=1200)
```

```python
# plot data with seaborn
sns.histplot(beta_sample, 
             color='#f67280',
             kde=True,
             stat='density')
plt.title('Beta Distribution', fontsize=15)
plt.show()
```

## 7 Geometric Distribution

* Models the number of trials that have to happen before it's a success
* Parameters: P - the probability of success of an individual trial

```python
geometric_sample = np.random.geometric(p=0.2, size=100)
# How many trials succeeded after a single run?
(geometric_sample == 1).sum()
```

```python
# plot data with seaborn
sns.histplot(geometric_sample, 
             color='#f67280',
             kde=True,
             stat='density')
plt.title('Geometric Distribution', fontsize=15)
plt.show()
```
